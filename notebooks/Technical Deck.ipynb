# Customer Segmentation Using Unsupervised Learning

**Objective**  
Identify interpretable and actionable customer segments using demographic and behavioral data.

**Why this matters**
- Customers are currently treated as homogeneous
- Inefficient promotion targeting and spend
- Need data-driven segmentation

**ML Task**
- Unsupervised Learning (Clustering)


# Dataset Overview

- Source: Kaggle â€“ Customer Purchases Behaviour Dataset
- Granularity: One row per customer
- No target variable (unsupervised)
- Mix of demographic, economic, and behavioral features


# Feature Categories

- **Demographic**: age, gender, education, region
- **Economic**: income
- **Behavioral**: purchase frequency, purchase amount, promotion usage
- **Relationship**: loyalty status, satisfaction score


# Data Cleaning & Preprocessing

**Key Decisions**
- Missing values: median (numeric), mode (categorical)
- Outliers: IQR-based winsorization
- Encoding: ordinal + one-hot
- Scaling: RobustScaler

**Why**
Distance-based clustering is sensitive to scale and noise.
